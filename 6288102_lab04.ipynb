# Exercise 2

Show to how plot graph of sin wave  5ùêªùëß , ampliture  $220\sqrt{2}$  from  0  second to  2  second

```python
import matplotlib.pyplot as plt
import numpy as np
import math
f = 5 #Hz
t = np.arange(0,2,0.02)
x = 220*math.sqrt(2)*np.sin(2*np.pi*f*t)
# print(x)
%matplotlib inline
plt.plot(t,x)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Simple sine (5 Hz)')
plt.ylim(-300,300)
plt.show()
```

How many complete cycles in one second?

What is V rms value of the signal?

What is average value of the signal?

```python
print("Complete cycles are 5 cycles")
print("V rms valuse of the signal is",np.std(x))
print("Average is",np.mean(x))
```

# Exercise 8

Show the matrix ($Z$) that can  translate vector $b$ to vector $x$

Proof  $Z*b =x$ using Python numpy

```python
Z = np.linalg.inv(A)
print(Z.dot(b))
np.array_equal(Z.dot(b),x)
```

```python
eigvals, eigvecs = np.linalg.eig(A)
Au = A.dot(eigvecs[:,0])
lambda_b = eigvals[0]*eigvecs[:,0]
```

```python
plt.quiver([0, 0], [0, 0], Au, lambda_b, angles='xy',scale_units='xy', scale=1,color=['r','b'])
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.grid(b=True, which='major') #<-- plot grid lines
```
